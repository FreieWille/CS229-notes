```python
import numpy as np
```

```python
class LogisticRegression:
    """
    逻辑回归

    模型假设：$y = g ( \theta' x ) = \frac{1}{1 + e^{- \theta' x}}$
    """
    def __init__(self, learning_rate=0.01, max_iterations=1000, fit_intercept=True, tol=1e-6):
        """
        参数:
        - learning_rate: 梯度下降学习率
        - max_iterations: 梯度下降最大迭代次数
        - fit_intercept: 截距是否为零
        - tol:梯度下降收敛容限(当损失变化小于tol时停止)
        """
        self.learning_rate = learning_rate
        self.max_iterations = max_iterations
        self.fit_intercept = fit_intercept
        self.tol = tol
        
        self.theta = None          # 需要拟合的向量
        self.loss_history = []     
```

```python
def _add_intercept(self, X):
    intercept = np.ones((X.shape[0], 1))
    return np.hstack((intercept, X)) # hstack()将具有相同行数的矩阵水平拼接

def _sigmoid(self, z):
    """
    sigmoid函数的数值稳定版本
    """
    z = np.clip(z, -500, 500) # numpy.clip()对输入做截断，大于最大值时返回最大值，小于最小值时返回最小值
    return 1.0 / (1.0 + np.exp(-z))

def _loss(self, h, y):
    """
    损失函数: 
    $\ell ( \theta ) = \sum_{i = 1}^n y^{(i)} \ln \left[ h_{\theta} ( x^{(i)} ) \right] + 
    ( 1 - y^{(i)} ) \ln \left[ 1 - h_{\theta} ( x^{(i)} ) \right]$，同时加入一个极小值防止出现log(0)的情况报错
    """
    eps = 1e-15  # 防止报错
    h = np.clip(h, eps, 1 - eps)
    return np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
```

```python
def fit(self, X, y):
    """
    使用梯度下降法

    参数:
    - X: 行数为样本数，列数等于向量$x^{(i)}$的长度的矩阵
    - y: 长度等于样本数且取值为0或1的列向量
    """
    # 截距不为零的情形
    if self.fit_intercept:
        X = self._add_intercept(X)

    m, d = X.shape
    self.theta = np.zeros((d, 1))  # 初始化 θ 为零向量

    for i in range(self.max_iterations):
        # 初始化
        z = X @ self.theta          
        h = self._sigmoid(z)        

        gradient = (X.T @ (h - y))

        self.theta = self.theta -  self.learning_rate * gradient # 更新步骤

        # 记录损失
        loss = self._loss(h, y)
        self.loss_history.append(loss)
        
    return self
```

```python
def predict_proba(self, X):
    """
    计算$h_{\theta} ( x )$
    """
    if self.fit_intercept:
        X = self._add_intercept(X)
    return self._sigmoid(X @ self.theta)

def predict(self, X, threshold=0.5):
    """
    根据阈值进行分类，阈值变量threshold(默认值为0.5) 
    """
    proba = self.predict_proba(X)
    return (proba >= threshold).astype(int).flatten() #astype()将bool类型转换为int类型，flatten()函数将二维数组转换为一维数组
```
